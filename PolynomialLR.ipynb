```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

Polynomial Linear Regression
Polynomial Linear Regression Genel Formülü:

y=a+b1x+b2x^2+b3x^3+ b4x^4 + ....... + bN*x^N

```python
df = pd.read_csv("/content/drive/MyDrive/ProjelerleYapayZMakineO/polynomial.csv", sep=";")
```

```python
polynomail_regression = PolynomialFeatures(degree = 4)
```

eğer biz bu resitene linear regresion uygularsal hiç uygun olmayan bir tahmin çizgisi görürüz

```python
plt.scatter(df['deneyim'], df['maas'])
plt.xlabel('deneyim(yıl)')
plt.ylabel('maas')
plt.show()
```

görüldüğü gibi doğrusal bir yapıda dağılmıyor veriler eğer biz bu resitene linear regresion uygularsal hiç uygun olmayan bir tahmin çizgisi görürüz

```python
reg = LinearRegression()
reg.fit(df[['deneyim']], df['maas'])

plt.xlabel('deneyim(yıl)')
plt.ylabel('maas')
plt.scatter(df['deneyim'], df['maas'])
xekseni = df['deneyim']
yekseni = reg.predict(df[['deneyim']])
plt.plot(xekseni, yekseni, color='red')
plt.show()
```

```python
polynomail_regression = PolynomialFeatures(degree = 2)
x_poly = polynomail_regression.fit_transform(df[['deneyim']])
```

```python
reg = LinearRegression()
reg.fit(x_poly, df['maas'])
```

```python
y_head = reg.predict(x_poly)
plt.plot(df['deneyim'], y_head, color='red', label = 'polynomial regression')
plt.legend()
plt.scatter(df['deneyim'], df['maas'])
plt.show()
```

gördüğümüz gibi kesinlikle uymuş diyebiliriz, polynomial regression kesinlikle uymuş diyebiliriz



```python
x_polynomial1 = polynomail_regression.fit_transform([[6.6]])
x_polynomial2 = polynomail_regression.fit_transform([[7.6]])
```
